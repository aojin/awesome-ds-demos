### Using groupby(), plot the number of films that have been released each decade in the history of cinema.

```python
c = cast
c.head()
```

```python
c.year = c.year//10*10
```

```python
c.groupby(['year']).count()['title']
```

### Use groupby() to plot the number of "Hamlet" films made each decade.

```python
ham = c[c.title == 'Hamlet']
```

```python
ham.groupby(['year']).count()['title']
```

### How many leading (n=1) roles were available to actors, and how many to actresses, in each year of the 1950s?

```python
lead = c[c.n == 1]
```

```python
lead.head()
```

```python
lead.groupby(['year']).count()
```

```python
lead.groupby(['year','type']).count()
```

### In the 1950s decade taken as a whole, how many total roles were available to actors, and how many to actresses, for each "n" number 1 through 5?

```python
fiftiesroles = c[c.year == 1950]
```

```python
fiftiesroles = fiftiesroles.reset_index('title')
```

```python
fiftiesroles.head()
```

```python
fiftiesroles.groupby('type').count()
```

```python
fiftiesroles[fiftiesroles.n < 5].groupby('type').count()
```
